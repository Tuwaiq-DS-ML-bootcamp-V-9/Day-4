### Task 1
1. Write a function named is_prime that takes a number as a parameter and returns True if the number is prime and False otherwise. Print whether the number 29 is prime.

```python
# write your code here ^_^
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

print("29 is prime:", is_prime(29))
```

### Task 2
1. Create a list of 10 numbers and sort them in descending order using the sort() function.
2. Use the len() function to print the length of the sorted list.
3. Use the zip() function to combine three lists (names, ages, and cities) into a list of tuples.
4. Use the range() function to create a list of even numbers from 2 to 20 and print the list.

```python
# write your code here ^_^

#step 1
numbers = [5, 2, 9, 1, 5, 6, 7, 3, 8, 4]
numbers.sort(reverse=True)
print("Sorted list in descending order:", numbers)

#step 2
print("Length of the sorted list:", len(numbers))

#step 3
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
cities = ["New York", "Los Angeles", "Chicago"]
combined = list(zip(names, ages, cities))
print("Combined list of tuples:", combined)

#step 4
even_numbers = list(range(2, 21, 2))
print("List of even numbers from 2 to 20:", even_numbers)
```

### Task 3
Build a phone book program that receives the phone number, and returns the name of the owner.
You can follow the table below:

| Name     | Number     |
| -------- | ---------- |
| Amal     | 1111111111 |
| Mohammed | 2222222222 |
| Khadijah | 3333333333 |
| Abdullah | 4444444444 |
| Rawan    | 5555555555 |
| Faisal   | 6666666666 |
| Layla    | 7777777777 |

- If the number exists, print the owner. Otherwise, print "Sorry, the number is not found".
- If the number is less or more than 10 numbers, print "This is invalid number".
- If the number contains letters or symbols, print "This is invalid number".


```python
# write your code here ^_^
phone_book = {
    "1111111111": "Amal",
    "2222222222": "Mohammed",
    "3333333333": "Khadijah",
    "4444444444": "Abdullah",
    "5555555555": "Rawan",
    "6666666666": "Faisal",
    "7777777777": "Layla"
}

def lookup_phone_number(phone_number):
    if not phone_number.isdigit() or len(phone_number) != 10:
        return "This is invalid number"
    return phone_book.get(phone_number, "Sorry, the number is not found")

# Example 
phone_number = input("Enter a phone number: ")
print(lookup_phone_number(phone_number))
```

### Task 4
- Given the following list : [5, 4, 17, 19, 30, 2, 7, 10, 45]
    1. Use a lambda function to filter out only the odd numbers.
    2. Print the new list of odd numbers.

```python
# write your code here ^_^

# step 1
original_list = [5, 4, 17, 19, 30, 2, 7, 10, 45]
odd_numbers = list(filter(lambda x: x % 2 != 0, original_list))

# step 2
print("List of odd numbers:", odd_numbers)
```
